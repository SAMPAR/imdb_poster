```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("MovieGenre.csv", encoding="ISO-8859-1")
df.head()
```

```python
df.info()
```

```python
def get_all_genres(genres):
    return df["Genre"].str.split("|", expand=True).stack().reset_index(level=1, drop=True).to_frame('genre')["genre"].unique()
    
all_genres = get_all_genres(df["Genre"])
all_genres
```

```python
genre_series = np.ndarray(shape=(df.shape[0], len(all_genres)))
genre_df = pd.DataFrame(genre_series, columns=all_genres)
df = df.append(genre_df)
df[all_genres] = df[all_genres].fillna(0).astype(int)
```

```python
genres = df["Genre"].str.split("|")
for idx, genre in enumerate(genres):
    if isinstance(genre, str):
        for g in genre:
            df.loc[idx, g] = 1
df
```
